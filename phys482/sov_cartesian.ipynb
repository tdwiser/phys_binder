```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib widget
```

```python
a, b = 1, 2
Nx, Ny = 100*a, 100*b

x = np.linspace(0,a,Nx)
y = np.linspace(0,b,Ny)

xx,yy = np.meshgrid(x,y)
```

```python
jmax = 100

j = np.arange(1,jmax+1)
```

```python
Aj = 2/b/np.sinh(j*np.pi*a/b)*b/j/np.pi*(1-(-1)**j)
```

```python
V = np.sum(Aj * np.sinh(np.pi*j*xx[...,None]/b) * np.sin(np.pi*j*yy[...,None]/b), axis=-1)
```

```python
V.shape
```

```python
fig,ax = plt.subplots()
ax.set_aspect('equal')
ax.contour(x,y,V)
```

```python
fig,ax = plt.subplots()
ax.plot(y,V[:,-1])
```

```python
fig,ax = plt.subplots()
ax.contourf(x,y,V)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_box_aspect((a,b,1))
ax.plot_surface(xx,yy,V)
```
